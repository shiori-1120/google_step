```python
def search_anagram(random_word, dictionary):
    sorted_ramdom_word = ''.join(sorted(random_word))
    
    new_dictionary = []
    for word in dictionary:
        new_dictionary.append((''.join(sorted(word)), word))
    sorted_new_dictionary = sorted(new_dictionary, key=lambda x: x[0])
    anagram = binary_search(sorted_ramdom_word, sorted_new_dictionary)
    
    return anagram
        
def binary_search(word, dictionary):
    left = 0
    right = len(dictionary) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if dictionary[mid][0] < word:
            left = mid + 1
        elif dictionary[mid][0] > word:
            right = mid - 1
        elif dictionary[mid][0] == word:
            searched_word_list = []
            searched_word_list.append(dictionary[mid][1])
            i = 1
            j = 1
            while dictionary[mid + i][0] == word:
                searched_word_list.append(dictionary[mid + i][1])
                i += 1
            while dictionary[mid - j][0] == word:
                searched_word_list.append(dictionary[mid - j][1])
                j += 1
            return searched_word_list
        
    return
```

```python
def search_anagrams(random_word_list, dictionary):
    sorted_ramdom_word_list = []
    for i in range(len(random_word_list)):
        sorted_ramdom_word_list.append(''.join(sorted(random_word_list[i])))
    
    new_dictionary = []
    for word in dictionary:
        new_dictionary.append((''.join(sorted(word)), word))
    sorted_new_dictionary = sorted(new_dictionary, key=lambda x: x[1])
    
    anagram_list = []
    for word in sorted_ramdom_word_list:
        anagram = binary_search(word, sorted_new_dictionary)
        if anagram != None:
            anagram_list.append(anagram)
    
    return anagram_list
```

```python
def upgrade_search_anagram(random_word, dictionary):
    ramdom_word_hashmap_list = []
    anagram_lists = []
    new_dictionary = []
    
    for word in random_word:
        ramdom_word_hashmap_list.append(make_alphabet_hashmap(word))

    for word in dictionary:
        new_dictionary.append([make_alphabet_hashmap(word), word])

    for hash_table in ramdom_word_hashmap_list:
        anagram_list = []
        for dictionary_hash_table in new_dictionary:
            for j in range(ord('a'), ord('z')+1):
                if hash_table[chr(j)] >= dictionary_hash_table[0][chr(j)]:
                    if j == ord('z'):
                        anagram_list.append(dictionary_hash_table[1])
                else: break
        anagram_lists.append(anagram_list)
    return anagram_lists

def make_alphabet_hashmap(word):
    hash_table = {}
    for i in range(ord('a'), ord('z')+1):
        hash_table[chr(i)] = word.count(chr(i))
        
    return hash_table
```

```python
def make_alphabet_hashmap(word):
    hash_table = {}
    for i in range(ord('a'), ord('z')):
        hash_table[chr(i)] = word.count(chr(i))
        
    return hash_table
```
